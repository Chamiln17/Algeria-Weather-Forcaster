```python
"""
Climate Change Analysis in Algeria - Phase 1: Data Preprocessing
================================================================
This script handles data loading, cleaning, monthly aggregation, and feature engineering.
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller, kpss
import warnings
warnings.filterwarnings('ignore')

# Set display options
pd.set_option('display.max_columns', None)
sns.set_style("whitegrid")

# ============================================================================
# STEP 1: DATA LOADING AND INITIAL EXPLORATION
# ============================================================================

def load_and_explore_data(filepath):
    """
    Load daily climate data and perform initial exploration.
    
    Parameters:
    -----------
    filepath : str
        Path to CSV file containing daily weather data
        
    Returns:
    --------
    df : pandas.DataFrame
        Loaded dataframe with parsed dates
    """
    print("=" * 70)
    print("STEP 1: LOADING DATA")
    print("=" * 70)
    
    # Load data
    df = pd.read_csv(filepath)
    
    # Parse dates and rename for clarity
    df['time'] = pd.to_datetime(df['time'])
    df = df.rename(columns={'time': 'date'})
    df = df.sort_values('date').reset_index(drop=True)
    
    print(f"\n✓ Loaded {len(df):,} daily records")
    print(f"✓ Date range: {df['date'].min()} to {df['date'].max()}")
    print(f"✓ Total years: {(df['date'].max() - df['date'].min()).days / 365.25:.1f}")
    
    # Display basic info
    print("\n" + "-" * 70)
    print("DATASET STRUCTURE:")
    print("-" * 70)
    print(df.info())
    
    print("\n" + "-" * 70)
    print("FIRST 5 ROWS:")
    print("-" * 70)
    print(df.head())
    
    return df


# ============================================================================
# STEP 2: MISSING VALUE ANALYSIS AND CLEANING
# ============================================================================

def analyze_missing_values(df):
    """
    Analyze missing values in the dataset.
    
    Parameters:
    -----------
    df : pandas.DataFrame
        Input dataframe
        
    Returns:
    --------
    missing_summary : pandas.DataFrame
        Summary of missing values by column
    """
    print("\n" + "=" * 70)
    print("STEP 2: MISSING VALUE ANALYSIS")
    print("=" * 70)
    
    missing = df.isnull().sum()
    missing_pct = 100 * missing / len(df)
    
    missing_summary = pd.DataFrame({
        'Column': missing.index,
        'Missing_Count': missing.values,
        'Missing_Percentage': missing_pct.values
    })
    
    missing_summary = missing_summary[missing_summary['Missing_Count'] > 0]
    missing_summary = missing_summary.sort_values('Missing_Count', ascending=False)
    
    if len(missing_summary) > 0:
        print("\n⚠️  Columns with missing values:")
        print(missing_summary.to_string(index=False))
    else:
        print("\n✓ No missing values detected!")
    
    return missing_summary


def clean_missing_values(df):
    """
    Handle missing values using appropriate strategies.
    
    Strategy:
    - Temperature: Linear interpolation (limit 5 days)
    - Precipitation: Fill with 0 (no rain assumed)
    - ET0: Linear interpolation (continuous process)
    - Wind/Radiation: Linear interpolation
    
    Parameters:
    -----------
    df : pandas.DataFrame
        Input dataframe with missing values
        
    Returns:
    --------
    df_clean : pandas.DataFrame
        Cleaned dataframe
    """
    print("\n" + "=" * 70)
    print("STEP 3: CLEANING MISSING VALUES")
    print("=" * 70)
    
    df_clean = df.copy()
    
    # Temperature variables: linear interpolation
    temp_cols = [col for col in df.columns if 'temperature' in col.lower()]
    for col in temp_cols:
        if col in df_clean.columns:
            before = df_clean[col].isnull().sum()
            df_clean[col] = df_clean[col].interpolate(method='linear', limit=5)
            after = df_clean[col].isnull().sum()
            print(f"✓ {col}: {before} → {after} missing values")
    
    # Precipitation: fill with 0
    precip_cols = [col for col in df.columns if 'precipitation' in col.lower() 
                   or 'rain' in col.lower() or 'snow' in col.lower()]
    for col in precip_cols:
        if col in df_clean.columns:
            before = df_clean[col].isnull().sum()
            df_clean[col] = df_clean[col].fillna(0)
            after = df_clean[col].isnull().sum()
            print(f"✓ {col}: {before} → {after} missing values (filled with 0)")
    
    # ET0 and other continuous variables: interpolation
    continuous_cols = [col for col in df.columns if 'et0' in col.lower() 
                      or 'wind' in col.lower() or 'radiation' in col.lower()]
    for col in continuous_cols:
        if col in df_clean.columns:
            before = df_clean[col].isnull().sum()
            df_clean[col] = df_clean[col].interpolate(method='linear', limit=5)
            after = df_clean[col].isnull().sum()
            print(f"✓ {col}: {before} → {after} missing values")
    
    return df_clean


# ============================================================================
# STEP 4: DAILY TO MONTHLY AGGREGATION
# ============================================================================

# In the aggregate_to_monthly function, replace the agg_rules dictionary:

def aggregate_to_monthly(df):
    """
    Aggregate daily data to monthly scale.
    """
    print("\n" + "=" * 70)
    print("STEP 4: MONTHLY AGGREGATION")
    print("=" * 70)
    
    # Set date as index
    df_monthly = df.set_index('date')
    
    # Define aggregation rules with ACTUAL column names from CSV
    agg_rules = {
        'temperature_2m_mean (°C)': 'mean',
        'temperature_2m_max (°C)': 'mean',
        'temperature_2m_min (°C)': 'mean',
        'precipitation_sum (mm)': 'sum',  # Total monthly precipitation
        'et0_fao_evapotranspiration (mm)': 'mean',
        'shortwave_radiation_sum (MJ/m²)': 'mean',
        'windspeed_10m_max (km/h)': 'mean'
    }
    
    # Keep only columns that exist
    agg_rules = {k: v for k, v in agg_rules.items() if k in df_monthly.columns}
    
    # Debug: Print which columns are being used
    print(f"\n✓ Columns found for aggregation: {list(agg_rules.keys())}")
    
    if not agg_rules:
        print("⚠️  No matching columns found! Available columns:")
        print(df_monthly.columns.tolist())
        raise ValueError("No columns match the aggregation rules")
    
    # Resample to monthly
    df_monthly = df_monthly.resample('MS').agg(agg_rules)
    
    # Rename columns to simpler names (optional but recommended)
    rename_map = {
        'temperature_2m_mean (°C)': 'temperature_2m_mean',
        'temperature_2m_max (°C)': 'temperature_2m_max',
        'temperature_2m_min (°C)': 'temperature_2m_min',
        'precipitation_sum (mm)': 'precipitation_sum',
        'et0_fao_evapotranspiration (mm)': 'et0_fao_evapotranspiration',
        'shortwave_radiation_sum (MJ/m²)': 'shortwave_radiation_sum',
        'windspeed_10m_max (km/h)': 'windspeed_10m_max'
    }
    df_monthly = df_monthly.rename(columns=rename_map)
    
    print(f"\n✓ Aggregated to {len(df_monthly)} monthly records")
    print(f"✓ Date range: {df_monthly.index.min()} to {df_monthly.index.max()}")
    print(f"\n" + "-" * 70)
    print("MONTHLY DATA PREVIEW:")
    print("-" * 70)
    print(df_monthly.head(10))
    
    return df_monthly

# ============================================================================
# STEP 5: FEATURE ENGINEERING
# ============================================================================

def engineer_features(df_monthly):
    """
    Create additional features for analysis.
    
    Features created:
    1. Temperature anomaly (deviation from baseline)
    2. Aridity Index (ET0 / Precipitation)
    3. Water Balance (Precipitation - ET0)
    4. Cumulative Water Balance
    5. Rolling statistics
    
    Parameters:
    -----------
    df_monthly : pandas.DataFrame
        Monthly dataframe
        
    Returns:
    --------
    df_features : pandas.DataFrame
        Dataframe with engineered features
    """
    print("\n" + "=" * 70)
    print("STEP 5: FEATURE ENGINEERING")
    print("=" * 70)
    
    df_features = df_monthly.copy()
    
    # 1. Temperature anomaly (deviation from mean)
    if 'temperature_2m_mean' in df_features.columns:
        baseline = df_features['temperature_2m_mean'].mean()
        df_features['temp_anomaly'] = df_features['temperature_2m_mean'] - baseline
        print(f"✓ Temperature anomaly (baseline: {baseline:.2f}°C)")
    
    # 2. Aridity Index (ET0 / Precipitation)
    if 'et0_fao_evapotranspiration' in df_features.columns and 'precipitation_sum' in df_features.columns:
        df_features['aridity_index'] = (
            df_features['et0_fao_evapotranspiration'] / 
            (df_features['precipitation_sum'] + 0.1)  # Avoid division by zero
        )
        print("✓ Aridity Index (ET0 / Precipitation)")
    
    # 3. Water Balance (P - ET0)
    if 'precipitation_sum' in df_features.columns and 'et0_fao_evapotranspiration' in df_features.columns:
        df_features['water_balance'] = (
            df_features['precipitation_sum'] - 
            df_features['et0_fao_evapotranspiration']
        )
        print("✓ Water Balance (Precipitation - ET0)")
    
    # 4. Cumulative Water Balance
    if 'water_balance' in df_features.columns:
        df_features['cumulative_water_balance'] = df_features['water_balance'].cumsum()
        print("✓ Cumulative Water Balance")
    
    # 5. Rolling statistics (12-month moving average)
    if 'temperature_2m_mean' in df_features.columns:
        df_features['temp_rolling_12m'] = df_features['temperature_2m_mean'].rolling(window=12).mean()
        print("✓ 12-month rolling temperature")
    
    if 'precipitation_sum' in df_features.columns:
        df_features['precip_rolling_12m'] = df_features['precipitation_sum'].rolling(window=12).sum()
        print("✓ 12-month rolling precipitation")
    
    print("\n" + "-" * 70)
    print("ENGINEERED FEATURES PREVIEW:")
    print("-" * 70)
    print(df_features[['temp_anomaly', 'aridity_index', 'water_balance']].describe())
    
    return df_features


# ============================================================================
# STEP 6: STATIONARITY TESTING
# ============================================================================

def test_stationarity(series, series_name):
    """
    Test if time series is stationary using ADF and KPSS tests.
    
    Stationarity: Statistical properties don't change over time.
    Required for ARIMA/SARIMA models.
    
    Tests:
    - ADF (Augmented Dickey-Fuller): H0 = non-stationary
    - KPSS: H0 = stationary
    
    Parameters:
    -----------
    series : pandas.Series
        Time series to test
    series_name : str
        Name of the series for display
        
    Returns:
    --------
    results : dict
        Dictionary containing test results
    """
    print(f"\n{'=' * 70}")
    print(f"STATIONARITY TEST: {series_name}")
    print('=' * 70)
    
    # Remove NaN values
    series_clean = series.dropna()
    
    # ADF Test
    adf_result = adfuller(series_clean, autolag='AIC')
    print("\n1. ADF Test (H0: Non-stationary)")
    print(f"   Test Statistic: {adf_result[0]:.4f}")
    print(f"   p-value: {adf_result[1]:.4f}")
    print(f"   Critical Values: {adf_result[4]}")
    
    if adf_result[1] < 0.05:
        print("   ✓ REJECT H0: Series is stationary (p < 0.05)")
        adf_stationary = True
    else:
        print("   ✗ FAIL TO REJECT H0: Series is non-stationary (p >= 0.05)")
        adf_stationary = False
    
    # KPSS Test
    kpss_result = kpss(series_clean, regression='ct')
    print("\n2. KPSS Test (H0: Stationary)")
    print(f"   Test Statistic: {kpss_result[0]:.4f}")
    print(f"   p-value: {kpss_result[1]:.4f}")
    print(f"   Critical Values: {kpss_result[3]}")
    
    if kpss_result[1] > 0.05:
        print("   ✓ FAIL TO REJECT H0: Series is stationary (p > 0.05)")
        kpss_stationary = True
    else:
        print("   ✗ REJECT H0: Series is non-stationary (p <= 0.05)")
        kpss_stationary = False
    
    # Conclusion
    print("\n" + "-" * 70)
    if adf_stationary and kpss_stationary:
        print("CONCLUSION: Series is STATIONARY")
        needs_differencing = False
    elif not adf_stationary and not kpss_stationary:
        print("CONCLUSION: Series is NON-STATIONARY")
        print("ACTION: Apply differencing for ARIMA/SARIMA")
        needs_differencing = True
    else:
        print("CONCLUSION: Tests disagree - examine series plot")
        needs_differencing = True
    
    return {
        'adf_statistic': adf_result[0],
        'adf_pvalue': adf_result[1],
        'kpss_statistic': kpss_result[0],
        'kpss_pvalue': kpss_result[1],
        'is_stationary': adf_stationary and kpss_stationary,
        'needs_differencing': needs_differencing
    }


# ============================================================================
# MAIN EXECUTION
# ============================================================================

if __name__ == "__main__":
    print("\n")
    print("█" * 70)
    print("  CLIMATE CHANGE ANALYSIS - ALGERIA")
    print("  Phase 1: Data Preprocessing Pipeline")
    print("█" * 70)
    
    # STEP 1: Load data
    # Replace 'algiers_weather_data.csv' with your actual file path
    filepath = 'Algiers_Weather_Data.csv'
    df = load_and_explore_data(filepath)
    
    # STEP 2-3: Analyze and clean missing values
    missing_summary = analyze_missing_values(df)
    df_clean = clean_missing_values(df)
    
    print(df_clean.head())
    # STEP 4: Aggregate to monthly
    df_monthly = aggregate_to_monthly(df_clean)
    
    # STEP 5: Engineer features
    df_features = engineer_features(df_monthly)
    
    # STEP 6: Test stationarity (key for SARIMA)
    if 'temperature_2m_mean' in df_features.columns:
        temp_stationarity = test_stationarity(
            df_features['temperature_2m_mean'], 
            'Monthly Temperature'
        )
    
    # Save processed data
    output_file = 'algiers_monthly_processed.csv'
    df_features.to_csv(output_file)
    print(f"\n{'=' * 70}")
    print(f"✓ PREPROCESSING COMPLETE!")
    print(f"✓ Saved to: {output_file}")
    print(f"{'=' * 70}\n")
    
    # Display final summary
    print("\n" + "=" * 70)
    print("FINAL DATASET SUMMARY")
    print("=" * 70)
    print(df_features.info())
    print("\n" + "-" * 70)
    print("STATISTICAL SUMMARY:")
    print("-" * 70)
    print(df_features.describe())
```
