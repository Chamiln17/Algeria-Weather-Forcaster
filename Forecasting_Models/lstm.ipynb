```python
"""
Climate Change Analysis - Phase 3: LSTM Forecasting
===================================================
LSTM = Long Short-Term Memory Neural Network

LSTMs are powerful for time series because they:
- Remember long-term patterns
- Handle seasonal cycles automatically
- Learn non-linear relationships
- Capture complex climate dynamics

Architecture: Input → LSTM Layers → Dropout → Dense Output
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import warnings
warnings.filterwarnings('ignore')

# TensorFlow/Keras imports
try:
    import tensorflow as tf
    from tensorflow import keras
    from keras.models import Sequential
    from keras.layers import LSTM, Dense, Dropout
    from keras.callbacks import EarlyStopping, ModelCheckpoint
    KERAS_AVAILABLE = True
    print(f"✓ TensorFlow version: {tf.__version__}")
except ImportError:
    KERAS_AVAILABLE = False
    print("⚠️  TensorFlow/Keras not installed!")
    print("   Install with: pip install tensorflow")

sns.set_style("whitegrid")

# ============================================================================
# STEP 1: LOAD DATA
# ============================================================================

def load_monthly_data(filepath):
    """Load monthly preprocessed data."""
    print("=" * 70)
    print("STEP 1: LOADING MONTHLY DATA")
    print("=" * 70)
    
    df = pd.read_csv(filepath, index_col=0, parse_dates=True)
    
    print(f"\n✓ Loaded {len(df)} monthly records")
    print(f"✓ Date range: {df.index.min()} to {df.index.max()}")
    
    return df


# ============================================================================
# STEP 2: PREPARE SEQUENCES FOR LSTM
# ============================================================================

def create_sequences(data, lookback=12):
    """
    Create sequences for LSTM training.
    
    LSTM needs sequences like:
    - Input: [Month 1, Month 2, ..., Month 12]
    - Output: Month 13
    
    This is called "supervised learning" transformation.
    
    Parameters:
    -----------
    data : numpy.array
        Scaled time series data
    lookback : int
        Number of past months to use for prediction
        
    Returns:
    --------
    X : numpy.array
        Input sequences (samples, lookback, features)
    y : numpy.array
        Target values (samples,)
    """
    print("\n" + "=" * 70)
    print("STEP 2: CREATING SEQUENCES")
    print("=" * 70)
    
    X, y = [], []
    
    for i in range(lookback, len(data)):
        # Input: previous 'lookback' months
        X.append(data[i-lookback:i])
        # Output: current month
        y.append(data[i])
    
    X = np.array(X)
    y = np.array(y)
    
    print(f"\n✓ Created {len(X)} sequences")
    print(f"✓ Input shape: {X.shape} (samples, lookback, features)")
    print(f"✓ Output shape: {y.shape}")
    
    # Reshape for LSTM: (samples, timesteps, features)
    X = X.reshape(X.shape[0], X.shape[1], 1)
    
    return X, y


# ============================================================================
# STEP 3: SCALE DATA (CRITICAL FOR NEURAL NETWORKS)
# ============================================================================

def scale_data(train, test):
    """
    Scale data to [0, 1] range using MinMaxScaler.
    
    Why? Neural networks learn better when inputs are normalized.
    
    Parameters:
    -----------
    train : pandas.Series
        Training data
    test : pandas.Series
        Test data
        
    Returns:
    --------
    train_scaled : numpy.array
        Scaled training data
    test_scaled : numpy.array
        Scaled test data
    scaler : MinMaxScaler
        Fitted scaler (needed for inverse transform)
    """
    print("\n" + "=" * 70)
    print("STEP 3: SCALING DATA")
    print("=" * 70)
    
    scaler = MinMaxScaler(feature_range=(0, 1))
    
    # Fit on training data only (avoid data leakage!)
    train_scaled = scaler.fit_transform(train.values.reshape(-1, 1))
    test_scaled = scaler.transform(test.values.reshape(-1, 1))
    
    print(f"\n✓ Training data scaled: {train_scaled.min():.4f} to {train_scaled.max():.4f}")
    print(f"✓ Test data scaled: {test_scaled.min():.4f} to {test_scaled.max():.4f}")
    print(f"✓ Original range: {train.min():.2f} to {train.max():.2f} °C")
    
    return train_scaled, test_scaled, scaler


# ============================================================================
# STEP 4: BUILD LSTM MODEL
# ============================================================================

def build_lstm_model(lookback, units_layer1=50, units_layer2=50, dropout=0.2):
    """
    Build LSTM neural network architecture.
    
    Architecture:
    - LSTM Layer 1: 50 neurons, return sequences
    - Dropout: 20% (prevents overfitting)
    - LSTM Layer 2: 50 neurons
    - Dropout: 20%
    - Dense Output: 1 neuron (temperature prediction)
    
    Parameters:
    -----------
    lookback : int
        Number of timesteps in input sequence
    units_layer1 : int
        Neurons in first LSTM layer
    units_layer2 : int
        Neurons in second LSTM layer
    dropout : float
        Dropout rate (0-1)
        
    Returns:
    --------
    model : keras.Sequential
        Compiled LSTM model
    """
    print("\n" + "=" * 70)
    print("STEP 4: BUILDING LSTM ARCHITECTURE")
    print("=" * 70)
    
    if not KERAS_AVAILABLE:
        raise ImportError("TensorFlow/Keras is required for LSTM")
    
    model = Sequential([
        # First LSTM layer
        LSTM(units=units_layer1, 
             activation='relu',
             return_sequences=True,  # Pass output to next LSTM
             input_shape=(lookback, 1)),
        
        # Dropout for regularization
        Dropout(dropout),
        
        # Second LSTM layer
        LSTM(units=units_layer2, 
             activation='relu',
             return_sequences=False),  # Final LSTM output
        
        # Dropout
        Dropout(dropout),
        
        # Output layer
        Dense(units=1)  # Single temperature value
    ])
    
    # Compile model
    model.compile(
        optimizer='adam',
        loss='mean_squared_error',
        metrics=['mae']
    )
    
    print("\n" + "-" * 70)
    print("MODEL ARCHITECTURE:")
    print("-" * 70)
    model.summary()
    
    return model


# ============================================================================
# STEP 5: TRAIN LSTM MODEL
# ============================================================================

def train_lstm(model, X_train, y_train, X_val, y_val, epochs=50, batch_size=16):
    """
    Train LSTM model with early stopping.
    
    Early stopping: Stop training if validation loss doesn't improve
    for 10 consecutive epochs (prevents overfitting).
    
    Parameters:
    -----------
    model : keras.Sequential
        LSTM model
    X_train, y_train : numpy.array
        Training sequences
    X_val, y_val : numpy.array
        Validation sequences
    epochs : int
        Maximum training epochs
    batch_size : int
        Batch size for training
        
    Returns:
    --------
    history : keras.History
        Training history
    """
    print("\n" + "=" * 70)
    print("STEP 5: TRAINING LSTM MODEL")
    print("=" * 70)
    
    # Callbacks
    early_stop = EarlyStopping(
        monitor='val_loss',
        patience=10,
        restore_best_weights=True,
        verbose=1
    )
    
    checkpoint = ModelCheckpoint(
        'best_lstm_model.h5',
        monitor='val_loss',
        save_best_only=True,
        verbose=0
    )
    
    print(f"\nTraining configuration:")
    print(f"- Epochs: {epochs}")
    print(f"- Batch size: {batch_size}")
    print(f"- Early stopping patience: 10")
    print(f"- Training samples: {len(X_train)}")
    print(f"- Validation samples: {len(X_val)}")
    print("\nTraining started...\n")
    
    # Train model
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=epochs,
        batch_size=batch_size,
        callbacks=[early_stop, checkpoint],
        verbose=1
    )
    
    print("\n✓ Training complete!")
    
    return history


# ============================================================================
# STEP 6: PLOT TRAINING HISTORY
# ============================================================================

def plot_training_history(history):
    """
    Visualize training and validation loss over epochs.
    
    Parameters:
    -----------
    history : keras.History
        Training history
    """
    print("\n" + "=" * 70)
    print("STEP 6: VISUALIZING TRAINING HISTORY")
    print("=" * 70)
    
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # Loss
    axes[0].plot(history.history['loss'], label='Training Loss', linewidth=2)
    axes[0].plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
    axes[0].set_xlabel('Epoch', fontsize=12)
    axes[0].set_ylabel('Loss (MSE)', fontsize=12)
    axes[0].set_title('Model Loss', fontsize=14, fontweight='bold')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)
    
    # MAE
    axes[1].plot(history.history['mae'], label='Training MAE', linewidth=2)
    axes[1].plot(history.history['val_mae'], label='Validation MAE', linewidth=2)
    axes[1].set_xlabel('Epoch', fontsize=12)
    axes[1].set_ylabel('MAE', fontsize=12)
    axes[1].set_title('Mean Absolute Error', fontsize=14, fontweight='bold')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('lstm_training_history.png', dpi=300, bbox_inches='tight')
    print("\n✓ Training history plot saved: lstm_training_history.png")
    plt.show()


# ============================================================================
# STEP 7: EVALUATE MODEL
# ============================================================================

def evaluate_lstm(model, X_test, y_test, scaler, test_dates):
    """
    Evaluate LSTM model on test set.
    
    Parameters:
    -----------
    model : keras.Sequential
        Trained LSTM model
    X_test, y_test : numpy.array
        Test sequences
    scaler : MinMaxScaler
        Fitted scaler for inverse transform
    test_dates : pandas.DatetimeIndex
        Dates corresponding to test set
        
    Returns:
    --------
    metrics : dict
        Evaluation metrics
    predictions : numpy.array
        Model predictions (unscaled)
    """
    print("\n" + "=" * 70)
    print("STEP 7: MODEL EVALUATION")
    print("=" * 70)
    
    # Make predictions (scaled)
    predictions_scaled = model.predict(X_test, verbose=0)
    
    # Inverse transform to original scale
    predictions = scaler.inverse_transform(predictions_scaled).flatten()
    y_test_unscaled = scaler.inverse_transform(y_test.reshape(-1, 1)).flatten()
    
    # Calculate metrics
    rmse = np.sqrt(mean_squared_error(y_test_unscaled, predictions))
    mae = mean_absolute_error(y_test_unscaled, predictions)
    r2 = r2_score(y_test_unscaled, predictions)
    
    metrics = {
        'RMSE': rmse,
        'MAE': mae,
        'R2': r2
    }
    
    print("\n" + "-" * 70)
    print("PERFORMANCE METRICS:")
    print("-" * 70)
    print(f"RMSE: {rmse:.4f} °C")
    print(f"MAE:  {mae:.4f} °C")
    print(f"R²:   {r2:.4f}")
    
    # Plot predictions vs actual
    fig, ax = plt.subplots(figsize=(14, 6))
    
    ax.plot(test_dates, y_test_unscaled, label='Actual', 
            color='green', linewidth=2, marker='o', markersize=4)
    ax.plot(test_dates, predictions, label='LSTM Predictions', 
            color='red', linewidth=2, linestyle='--')
    
    ax.set_xlabel('Date', fontsize=12)
    ax.set_ylabel('Temperature (°C)', fontsize=12)
    ax.set_title('LSTM Model: Predictions vs Actual', 
                 fontsize=14, fontweight='bold')
    ax.legend(loc='best', fontsize=10)
    ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('lstm_evaluation.png', dpi=300, bbox_inches='tight')
    print("\n✓ Evaluation plot saved: lstm_evaluation.png")
    plt.show()
    
    return metrics, predictions


# ============================================================================
# STEP 8: FORECAST FUTURE (ITERATIVE)
# ============================================================================

def forecast_future_lstm(model, last_sequence, scaler, months_ahead, last_date):
    """
    Forecast future temperatures iteratively.
    
    Process:
    1. Predict next month using last 12 months
    2. Add prediction to sequence
    3. Remove oldest month
    4. Repeat
    
    Parameters:
    -----------
    model : keras.Sequential
        Trained LSTM model
    last_sequence : numpy.array
        Last sequence from training data (scaled)
    scaler : MinMaxScaler
        Fitted scaler
    months_ahead : int
        Number of months to forecast
    last_date : datetime
        Last date in training data
        
    Returns:
    --------
    forecast_df : pandas.DataFrame
        Future forecasts with dates
    """
    print("\n" + "=" * 70)
    print("STEP 8: FORECASTING FUTURE")
    print("=" * 70)
    
    print(f"\nForecasting {months_ahead} months ahead...")
    
    # Initialize
    current_sequence = last_sequence.copy()
    forecasts = []
    
    # Iterative forecasting
    for i in range(months_ahead):
        # Predict next month
        next_pred_scaled = model.predict(current_sequence.reshape(1, -1, 1), verbose=0)
        forecasts.append(next_pred_scaled[0, 0])
        
        # Update sequence: remove oldest, add newest
        current_sequence = np.append(current_sequence[1:], next_pred_scaled[0, 0])
    
    # Inverse transform
    forecasts_unscaled = scaler.inverse_transform(
        np.array(forecasts).reshape(-1, 1)
    ).flatten()
    
    # Create dates
    forecast_dates = pd.date_range(
        start=last_date + pd.DateOffset(months=1),
        periods=months_ahead,
        freq='MS'
    )
    
    # Create DataFrame
    forecast_df = pd.DataFrame({
        'Date': forecast_dates,
        'Forecast': forecasts_unscaled
    })
    forecast_df.set_index('Date', inplace=True)
    
    print("\n✓ Forecast complete!")
    print("\n" + "-" * 70)
    print("FORECAST FOR 2040:")
    print("-" * 70)
    forecast_2040 = forecast_df[forecast_df.index.year == 2040]
    if len(forecast_2040) > 0:
        avg_2040 = forecast_2040['Forecast'].mean()
        print(f"Average Temperature: {avg_2040:.2f} °C")
        print("\nMonthly values:")
        print(forecast_2040)
    
    return forecast_df


def plot_long_term_forecast_lstm(train_data, forecast_df):
    """
    Plot long-term LSTM forecast.
    
    Parameters:
    -----------
    train_data : pandas.Series
        Historical training data
    forecast_df : pandas.DataFrame
        Forecast results
    """
    fig, ax = plt.subplots(figsize=(16, 6))
    
    # Historical data
    ax.plot(train_data.index, train_data, label='Historical Data', 
            color='steelblue', linewidth=2)
    
    # Forecast
    ax.plot(forecast_df.index, forecast_df['Forecast'], 
            label='LSTM Forecast', color='red', linewidth=2, linestyle='--')
    
    ax.axvline(train_data.index[-1], color='black', linestyle=':', 
               linewidth=2, label='Forecast Start')
    
    ax.set_xlabel('Date', fontsize=12)
    ax.set_ylabel('Temperature (°C)', fontsize=12)
    ax.set_title('LSTM Long-Term Temperature Forecast (2024-2040)', 
                 fontsize=14, fontweight='bold')
    ax.legend(loc='best', fontsize=10)
    ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('lstm_long_term_forecast.png', dpi=300, bbox_inches='tight')
    print("\n✓ Long-term forecast plot saved: lstm_long_term_forecast.png")
    plt.show()


# ============================================================================
# MAIN EXECUTION
# ============================================================================

if __name__ == "__main__":
    print("\n")
    print("█" * 70)
    print("  LSTM TEMPERATURE FORECASTING")
    print("  Climate Change Analysis - Algeria")
    print("█" * 70)
    
    if not KERAS_AVAILABLE:
        print("\n⚠️  TensorFlow/Keras not available. Cannot run LSTM.")
        print("   Install with: pip install tensorflow")
        exit()
    
    # Configuration
    LOOKBACK = 12  # Use 12 months to predict next month
    EPOCHS = 50
    BATCH_SIZE = 16
    TARGET_YEAR = 2040
    
    # STEP 1: Load data
    filepath = 'algiers_monthly_processed.csv'
    df = load_monthly_data(filepath)
    
    temp_series = df['temperature_2m_mean'].dropna()
    
    # STEP 2: Train/Validation/Test split
    # 70% train, 10% validation, 20% test
    total_len = len(temp_series)
    train_len = int(total_len * 0.7)
    val_len = int(total_len * 0.1)
    
    train = temp_series[:train_len]
    val = temp_series[train_len:train_len+val_len]
    test = temp_series[train_len+val_len:]
    
    print("\n" + "=" * 70)
    print("DATA SPLIT")
    print("=" * 70)
    print(f"Training:   {len(train)} months ({train.index[0]} to {train.index[-1]})")
    print(f"Validation: {len(val)} months ({val.index[0]} to {val.index[-1]})")
    print(f"Testing:    {len(test)} months ({test.index[0]} to {test.index[-1]})")
    
    # STEP 3: Scale data
    train_scaled, _, scaler = scale_data(train, test)
    val_scaled = scaler.transform(val.values.reshape(-1, 1))
    test_scaled = scaler.transform(test.values.reshape(-1, 1))
    
    # STEP 4: Create sequences
    X_train, y_train = create_sequences(train_scaled, lookback=LOOKBACK)
    X_val, y_val = create_sequences(val_scaled, lookback=LOOKBACK)
    X_test, y_test = create_sequences(test_scaled, lookback=LOOKBACK)
    
    # Adjust test dates (account for lookback)
    test_dates = test.index[LOOKBACK:]
    
    # STEP 5: Build model
    model = build_lstm_model(lookback=LOOKBACK, units_layer1=50, units_layer2=50)
    
    # STEP 6: Train model
    history = train_lstm(model, X_train, y_train, X_val, y_val, 
                        epochs=EPOCHS, batch_size=BATCH_SIZE)
    
    # STEP 7: Plot training history
    plot_training_history(history)
    
    # STEP 8: Evaluate on test set
    metrics, predictions = evaluate_lstm(model, X_test, y_test, scaler, test_dates)
    
    # STEP 9: Forecast to 2040
    last_sequence = train_scaled[-LOOKBACK:]
    months_ahead = (TARGET_YEAR - train.index[-1].year) * 12 + (12 - train.index[-1].month)
    
    forecast_df = forecast_future_lstm(
        model, last_sequence, scaler, months_ahead, train.index[-1]
    )
    plot_long_term_forecast_lstm(train, forecast_df)
    
    # Save results
    forecast_df.to_csv('lstm_forecast_2040.csv')
    model.save('lstm_temperature_model.h5')
    
    print("\n" + "=" * 70)
    print("✓ LSTM ANALYSIS COMPLETE!")
    print("=" * 70)
    print(f"✓ Forecast saved: lstm_forecast_2040.csv")
    print(f"✓ Model saved: lstm_temperature_model.h5")
    print(f"✓ Performance: RMSE={metrics['RMSE']:.4f}, R²={metrics['R2']:.4f}")
    print("=" * 70 + "\n")
```
