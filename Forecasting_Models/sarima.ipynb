```python
"""
Climate Change Analysis - Phase 2: SARIMA Forecasting
=====================================================
SARIMA = Seasonal AutoRegressive Integrated Moving Average

This model captures:
- Trend (I: Integrated)
- Seasonality (S: Seasonal)
- Autocorrelation (AR: AutoRegressive)
- Moving Average (MA)

Perfect for temperature data with annual cycles!
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import warnings
warnings.filterwarnings('ignore')

# Attempt to import pmdarima for automatic order selection
try:
    from pmdarima import auto_arima
    PMDARIMA_AVAILABLE = True
except ImportError:
    PMDARIMA_AVAILABLE = False
    print("⚠️  pmdarima not installed. Using manual order selection.")
    print("   Install with: pip install pmdarima")

sns.set_style("whitegrid")

# ============================================================================
# STEP 1: LOAD PREPROCESSED DATA
# ============================================================================

def load_monthly_data(filepath):
    """
    Load monthly preprocessed data from Phase 1.
    
    Parameters:
    -----------
    filepath : str
        Path to monthly CSV file
        
    Returns:
    --------
    df : pandas.DataFrame
        Monthly dataframe with datetime index
    """
    print("=" * 70)
    print("STEP 1: LOADING MONTHLY DATA")
    print("=" * 70)
    
    df = pd.read_csv(filepath, index_col=0, parse_dates=True)
    
    print(f"\n✓ Loaded {len(df)} monthly records")
    print(f"✓ Date range: {df.index.min()} to {df.index.max()}")
    print(f"\n{df.head()}")
    
    return df


# ============================================================================
# STEP 2: VISUALIZE SEASONAL DECOMPOSITION
# ============================================================================

def decompose_series(series, period=12):
    """
    Decompose time series into trend, seasonal, and residual components.
    
    This helps understand:
    - Long-term trend (climate change signal)
    - Seasonal pattern (summer/winter cycles)
    - Random noise (weather variability)
    
    Parameters:
    -----------
    series : pandas.Series
        Time series to decompose
    period : int
        Seasonal period (12 for monthly data)
        
    Returns:
    --------
    decomposition : DecomposeResult
        Object containing trend, seasonal, residual components
    """
    print("\n" + "=" * 70)
    print("STEP 2: SEASONAL DECOMPOSITION")
    print("=" * 70)
    
    # Remove NaN values
    series_clean = series.dropna()
    
    # Decompose (additive model: Y = Trend + Seasonal + Residual)
    decomposition = seasonal_decompose(
        series_clean,
        model='additive',
        period=period,
        extrapolate_trend='freq'
    )
    
    # Plot
    fig, axes = plt.subplots(4, 1, figsize=(12, 10))
    
    # Original
    axes[0].plot(series_clean, color='steelblue', linewidth=1.5)
    axes[0].set_ylabel('Original', fontsize=12)
    axes[0].set_title('Seasonal Decomposition of Temperature', fontsize=14, fontweight='bold')
    axes[0].grid(True, alpha=0.3)
    
    # Trend
    axes[1].plot(decomposition.trend, color='darkred', linewidth=2)
    axes[1].set_ylabel('Trend', fontsize=12)
    axes[1].grid(True, alpha=0.3)
    
    # Seasonal
    axes[2].plot(decomposition.seasonal, color='green', linewidth=1.5)
    axes[2].set_ylabel('Seasonal', fontsize=12)
    axes[2].grid(True, alpha=0.3)
    
    # Residual
    axes[3].plot(decomposition.resid, color='purple', linewidth=1)
    axes[3].set_ylabel('Residual', fontsize=12)
    axes[3].set_xlabel('Date', fontsize=12)
    axes[3].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('seasonal_decomposition.png', dpi=300, bbox_inches='tight')
    print("\n✓ Decomposition plot saved: seasonal_decomposition.png")
    plt.show()
    
    # Print insights
    print("\n" + "-" * 70)
    print("INSIGHTS:")
    print("-" * 70)
    print(f"Trend range: {decomposition.trend.min():.2f} to {decomposition.trend.max():.2f}°C")
    print(f"Seasonal amplitude: {decomposition.seasonal.max() - decomposition.seasonal.min():.2f}°C")
    print(f"Residual std: {decomposition.resid.std():.2f}°C")
    
    return decomposition


# ============================================================================
# STEP 3: ACF/PACF PLOTS (Understand autocorrelation)
# ============================================================================

def plot_acf_pacf(series, lags=40):
    """
    Plot ACF and PACF to help identify SARIMA orders.
    
    ACF: Shows correlation with past values
    PACF: Shows direct correlation (removing intermediate effects)
    
    Parameters:
    -----------
    series : pandas.Series
        Time series
    lags : int
        Number of lags to plot
    """
    print("\n" + "=" * 70)
    print("STEP 3: ACF/PACF ANALYSIS")
    print("=" * 70)
    
    series_clean = series.dropna()
    
    fig, axes = plt.subplots(2, 1, figsize=(12, 8))
    
    # ACF
    plot_acf(series_clean, lags=lags, ax=axes[0])
    axes[0].set_title('Autocorrelation Function (ACF)', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Lag', fontsize=12)
    
    # PACF
    plot_pacf(series_clean, lags=lags, ax=axes[1])
    axes[1].set_title('Partial Autocorrelation Function (PACF)', fontsize=14, fontweight='bold')
    axes[1].set_xlabel('Lag', fontsize=12)
    
    plt.tight_layout()
    plt.savefig('acf_pacf_plots.png', dpi=300, bbox_inches='tight')
    print("\n✓ ACF/PACF plot saved: acf_pacf_plots.png")
    plt.show()
    
    print("\nInterpretation Guide:")
    print("- ACF cuts off after lag q → MA(q)")
    print("- PACF cuts off after lag p → AR(p)")
    print("- Both decay gradually → ARMA(p,q)")


# ============================================================================
# STEP 4: AUTOMATIC SARIMA ORDER SELECTION
# ============================================================================

def find_best_sarima_order(train_data, seasonal_period=12):
    """
    Use auto_arima to find optimal SARIMA orders.
    
    SARIMA(p,d,q)(P,D,Q,s):
    - p: AR order (non-seasonal)
    - d: Differencing order (non-seasonal)
    - q: MA order (non-seasonal)
    - P: Seasonal AR order
    - D: Seasonal differencing
    - Q: Seasonal MA order
    - s: Seasonal period (12 for monthly)
    
    Parameters:
    -----------
    train_data : pandas.Series
        Training data
    seasonal_period : int
        Seasonal period (12 for monthly)
        
    Returns:
    --------
    best_order : tuple
        (p,d,q)
    best_seasonal_order : tuple
        (P,D,Q,s)
    """
    print("\n" + "=" * 70)
    print("STEP 4: AUTOMATIC ORDER SELECTION")
    print("=" * 70)
    
    if not PMDARIMA_AVAILABLE:
        print("\n⚠️  Using default order: SARIMA(1,1,1)(1,1,1,12)")
        return (1, 1, 1), (1, 1, 1, 12)
    
    print("\nSearching for optimal SARIMA orders...")
    print("This may take 2-5 minutes...\n")
    
    model = auto_arima(
        train_data,
        start_p=0, start_q=0,
        max_p=5, max_q=5,
        start_P=0, start_Q=0,
        max_P=2, max_Q=2,
        seasonal=True,
        m=seasonal_period,
        d=None,  # Let auto_arima determine
        D=None,  # Let auto_arima determine
        trace=True,
        error_action='ignore',
        suppress_warnings=True,
        stepwise=True
    )
    
    print("\n" + "=" * 70)
    print("✓ BEST MODEL FOUND")
    print("=" * 70)
    print(f"Order: {model.order}")
    print(f"Seasonal Order: {model.seasonal_order}")
    print(f"AIC: {model.aic():.2f}")
    print(f"BIC: {model.bic():.2f}")
    
    return model.order, model.seasonal_order


# ============================================================================
# STEP 5: TRAIN SARIMA MODEL
# ============================================================================

def train_sarima(train_data, order, seasonal_order):
    """
    Train SARIMA model with specified orders.
    
    Parameters:
    -----------
    train_data : pandas.Series
        Training data
    order : tuple
        (p,d,q)
    seasonal_order : tuple
        (P,D,Q,s)
        
    Returns:
    --------
    model_fit : SARIMAXResults
        Fitted SARIMA model
    """
    print("\n" + "=" * 70)
    print("STEP 5: TRAINING SARIMA MODEL")
    print("=" * 70)
    
    print(f"\nModel: SARIMA{order}x{seasonal_order}")
    print("Fitting model...")
    
    model = SARIMAX(
        train_data,
        order=order,
        seasonal_order=seasonal_order,
        enforce_stationarity=False,
        enforce_invertibility=False
    )
    
    model_fit = model.fit(disp=False)
    
    print("\n✓ Model trained successfully!")
    print("\n" + "-" * 70)
    print("MODEL SUMMARY:")
    print("-" * 70)
    print(model_fit.summary())
    
    return model_fit


# ============================================================================
# STEP 6: EVALUATE MODEL PERFORMANCE
# ============================================================================

def evaluate_model(model_fit, train_data, test_data):
    """
    Evaluate SARIMA model on test set.
    
    Metrics:
    - RMSE: Root Mean Squared Error (lower is better)
    - MAE: Mean Absolute Error (lower is better)
    - R²: Coefficient of determination (higher is better, 0-1)
    
    Parameters:
    -----------
    model_fit : SARIMAXResults
        Fitted model
    train_data : pandas.Series
        Training data
    test_data : pandas.Series
        Test data
        
    Returns:
    --------
    metrics : dict
        Dictionary of evaluation metrics
    predictions : pandas.Series
        Model predictions on test set
    """
    print("\n" + "=" * 70)
    print("STEP 6: MODEL EVALUATION")
    print("=" * 70)
    
    # Get predictions
    predictions = model_fit.get_forecast(steps=len(test_data))
    pred_mean = predictions.predicted_mean
    pred_ci = predictions.conf_int(alpha=0.05)
    
    # Calculate metrics
    rmse = np.sqrt(mean_squared_error(test_data, pred_mean))
    mae = mean_absolute_error(test_data, pred_mean)
    r2 = r2_score(test_data, pred_mean)
    
    metrics = {
        'RMSE': rmse,
        'MAE': mae,
        'R2': r2
    }
    
    print("\n" + "-" * 70)
    print("PERFORMANCE METRICS:")
    print("-" * 70)
    print(f"RMSE: {rmse:.4f} °C")
    print(f"MAE:  {mae:.4f} °C")
    print(f"R²:   {r2:.4f}")
    
    # Plot results
    fig, ax = plt.subplots(figsize=(14, 6))
    
    # Training data
    ax.plot(train_data.index, train_data, label='Training Data', 
            color='steelblue', linewidth=2)
    
    # Test data
    ax.plot(test_data.index, test_data, label='Test Data (Actual)', 
            color='green', linewidth=2, marker='o', markersize=4)
    
    # Predictions
    ax.plot(pred_mean.index, pred_mean, label='SARIMA Predictions', 
            color='red', linewidth=2, linestyle='--')
    
    # Confidence interval
    ax.fill_between(pred_ci.index, 
                    pred_ci.iloc[:, 0], 
                    pred_ci.iloc[:, 1],
                    color='red', alpha=0.2, label='95% Confidence Interval')
    
    ax.set_xlabel('Date', fontsize=12)
    ax.set_ylabel('Temperature (°C)', fontsize=12)
    ax.set_title('SARIMA Model: Training vs Test Performance', 
                 fontsize=14, fontweight='bold')
    ax.legend(loc='best', fontsize=10)
    ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('sarima_evaluation.png', dpi=300, bbox_inches='tight')
    print("\n✓ Evaluation plot saved: sarima_evaluation.png")
    plt.show()
    
    return metrics, pred_mean, pred_ci


# ============================================================================
# STEP 7: FORECAST FUTURE (2024-2040)
# ============================================================================

def forecast_future(model_fit, last_date, target_year=2040):
    """
    Forecast temperature to target year.
    
    Parameters:
    -----------
    model_fit : SARIMAXResults
        Fitted model
    last_date : datetime
        Last date in training data
    target_year : int
        Year to forecast to
        
    Returns:
    --------
    forecast_df : pandas.DataFrame
        Forecast with confidence intervals
    """
    print("\n" + "=" * 70)
    print(f"STEP 7: FORECASTING TO {target_year}")
    print("=" * 70)
    
    # Calculate number of months to forecast
    months_ahead = (target_year - last_date.year) * 12 + (12 - last_date.month)
    
    print(f"\nForecasting {months_ahead} months ahead...")
    print(f"From {last_date.strftime('%Y-%m')} to {target_year}-12")
    
    # Get forecast
    forecast = model_fit.get_forecast(steps=months_ahead)
    forecast_mean = forecast.predicted_mean
    forecast_ci = forecast.conf_int(alpha=0.05)
    
    # Create DataFrame
    forecast_df = pd.DataFrame({
        'Forecast': forecast_mean,
        'Lower_CI': forecast_ci.iloc[:, 0],
        'Upper_CI': forecast_ci.iloc[:, 1]
    })
    
    print("\n" + "-" * 70)
    print(f"FORECAST FOR {target_year}:")
    print("-" * 70)
    
    # Get 2040 forecast (last 12 months)
    forecast_2040 = forecast_df.tail(12)
    avg_2040 = forecast_2040['Forecast'].mean()
    ci_lower = forecast_2040['Lower_CI'].mean()
    ci_upper = forecast_2040['Upper_CI'].mean()
    
    print(f"Average Temperature: {avg_2040:.2f} °C")
    print(f"95% CI: [{ci_lower:.2f}, {ci_upper:.2f}] °C")
    
    print("\n" + "-" * 70)
    print("MONTHLY FORECAST FOR 2040:")
    print("-" * 70)
    print(forecast_2040[['Forecast', 'Lower_CI', 'Upper_CI']])
    
    return forecast_df


def plot_long_term_forecast(train_data, forecast_df):
    """
    Plot long-term forecast visualization.
    
    Parameters:
    -----------
    train_data : pandas.Series
        Historical training data
    forecast_df : pandas.DataFrame
        Forecast results
    """
    fig, ax = plt.subplots(figsize=(16, 6))
    
    # Historical data
    ax.plot(train_data.index, train_data, label='Historical Data', 
            color='steelblue', linewidth=2)
    
    # Forecast
    ax.plot(forecast_df.index, forecast_df['Forecast'], 
            label='SARIMA Forecast', color='red', linewidth=2, linestyle='--')
    
    # Confidence interval
    ax.fill_between(forecast_df.index,
                    forecast_df['Lower_CI'],
                    forecast_df['Upper_CI'],
                    color='red', alpha=0.2, label='95% Confidence Interval')
    
    ax.axvline(train_data.index[-1], color='black', linestyle=':', 
               linewidth=2, label='Forecast Start')
    
    ax.set_xlabel('Date', fontsize=12)
    ax.set_ylabel('Temperature (°C)', fontsize=12)
    ax.set_title('SARIMA Long-Term Temperature Forecast (2024-2040)', 
                 fontsize=14, fontweight='bold')
    ax.legend(loc='best', fontsize=10)
    ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('sarima_long_term_forecast.png', dpi=300, bbox_inches='tight')
    print("\n✓ Long-term forecast plot saved: sarima_long_term_forecast.png")
    plt.show()


# ============================================================================
# MAIN EXECUTION
# ============================================================================

if __name__ == "__main__":
    print("\n")
    print("█" * 70)
    print("  SARIMA TEMPERATURE FORECASTING")
    print("  Climate Change Analysis - Algeria")
    print("█" * 70)
    
    # STEP 1: Load data
    filepath = 'algiers_monthly_processed.csv'
    df = load_monthly_data(filepath)
    
    # Select temperature column
    temp_series = df['temperature_2m_mean'].dropna()
    
    # STEP 2: Seasonal decomposition
    decomp = decompose_series(temp_series, period=12)
    
    # STEP 3: ACF/PACF plots
    plot_acf_pacf(temp_series, lags=40)
    
    # STEP 4: Train/Test split (80/20)
    split_idx = int(len(temp_series) * 0.8)
    train = temp_series[:split_idx]
    test = temp_series[split_idx:]
    
    print("\n" + "=" * 70)
    print("DATA SPLIT")
    print("=" * 70)
    print(f"Training: {len(train)} months ({train.index[0]} to {train.index[-1]})")
    print(f"Testing:  {len(test)} months ({test.index[0]} to {test.index[-1]})")
    
    # STEP 5: Find best SARIMA order
    order, seasonal_order = find_best_sarima_order(train, seasonal_period=12)
    
    # STEP 6: Train model
    model_fit = train_sarima(train, order, seasonal_order)
    
    # STEP 7: Evaluate on test set
    metrics, predictions, pred_ci = evaluate_model(model_fit, train, test)
    
    # STEP 8: Forecast to 2040
    forecast_df = forecast_future(model_fit, train.index[-1], target_year=2040)
    plot_long_term_forecast(train, forecast_df)
    
    # Save results
    forecast_df.to_csv('sarima_forecast_2040.csv')
    print("\n" + "=" * 70)
    print("✓ SARIMA ANALYSIS COMPLETE!")
    print("=" * 70)
    print(f"✓ Forecast saved: sarima_forecast_2040.csv")
    print(f"✓ Model Performance: RMSE={metrics['RMSE']:.4f}, R²={metrics['R2']:.4f}")
    print("=" * 70 + "\n")
```
